# Japanese cosmetic company brand  trend comparison

```python
all_period_brands = pd.read_csv('input/cosmetic_company.csv')
key_brands = 'HAKU'
brands_name= []
for i in range(0, len(all_period_brands)):
    #brand2がnanでない場合
    if isinstance(all_period_brands.iloc[i][4], str) == 1:
        first_second_brands = list(all_period_brands.iloc[i][3:5])
        if key_brands not in first_second_brands:
            first_second_brands+= [key_brands]
            brands_name.append(first_second_brands)
            #brand3がnanでない場合
            if isinstance(all_period_brands.iloc[i][5], str) == 1:
                third_brands = [all_period_brands.iloc[i][5]]
                
                if key_brands not in third_brands:
                    third_brands+= [key_brands]
                    brands_name.append(third_brands)
                else:
                    brands_name.append(third_brands)
            
        else:
            brands_name.append(first_second_brands)
            if isinstance(all_period_brands.iloc[i][5], str) == 1:
                third_brands = [all_period_brands.iloc[i][5]]
                
                if key_brands not in third_brands:
                    third_brands+= [key_brands]
                    brands_name.append(third_brands)
                else:
                    brands_name.append(third_brands)
   
                               
    #brand2がnanである場合   
    else:
        first_second_brands =  all_period_brands.iloc[i][3]
        if key_brands not in first_second_brands:
            set = {first_second_brands, key_brands}
            first_second_brands = list(set)
            brands_name.append(first_second_brands) 
        else:
            brands_name.append(first_second_brands)  

brands_name_list = []
trend_time = '2007-09-01 2020-07-08'
pytrends = TrendReq(hl='ja-JP', tz= -450)

for kw_list in brands_name:
    pytrends.build_payload(kw_list, timeframe=trend_time, geo='JP')
    df = pytrends.interest_over_time()
    brands_name_list.append(df)
    if kw_list ==  brands_name[0]:
        df_final = df
    else:
        df_final = pd.concat([df_final, df], axis=1)

brand1 = df_final.T.reset_index().rename(columns = {'index': 'brand1'})
brand1.drop_duplicates(subset='brand1', keep='first', inplace= True)
brand2 = brand1.rename(columns = {'brand1': 'brand2'})
brand3 = brand1.rename(columns = {'brand1': 'brand3'})
to_get_brands = pd.read_csv('input/cosmetic_company.csv')
temp_1 = to_get_brands[['company_id', 'brand1','Date']].dropna(subset=['brand1'])
temp_2 = to_get_brands[['company_id', 'brand2','Date']].dropna(subset=['brand2'])
temp_3 = to_get_brands[['company_id', 'brand3','Date']].dropna(subset=['brand3'])

temp_1 = pd.merge(temp_1, brand1, on=['brand1'], how='left')
temp_2 = pd.merge(temp_2, brand2, on=['brand2'], how='left')
temp_3 = pd.merge(temp_3, brand3, on=['brand3'], how='left')

brands_trend = pd.concat([temp_1, temp_2, temp_3], axis=0)
brands_trend = brands_trend.fillna(' ')
brands_trend['brandall'] = brands_trend['brand1'] + brands_trend['brand2'] + brands_trend['brand3'] 
brands_trend['brandall'] = brands_trend['brandall'].str.strip()
brands_trend.drop(columns=['brand1','brand2', 'brand3'], inplace = True)
brands_trend.reset_index(inplace = True , drop=True)

brands_trend.to_pickle('brands_trend')

after_con = brands_trend[['company_id', 'brandall','Date']]
num_brands_trend = brands_trend.T
num_brands_trend.iloc[2:-1]
num_brands_trend.iloc[2:-1].index = pd.to_datetime(num_brands_trend.iloc[2:-1].index)
brands_trend_12 = num_brands_trend.iloc[2:-1].shift(1).rolling(window=12).agg(['median', 'max', 'sum', 'std', 'mean'])
brands_trend_6 = num_brands_trend.iloc[2:-1].shift(1).rolling(window=6).agg(['median', 'max', 'sum', 'std', 'mean'])
brands_trend_24 = num_brands_trend.iloc[2:-1].shift(1).rolling(window=24).agg(['median', 'max', 'sum', 'std', 'mean'])


##12ヶ月
brands_trend_12.T.reset_index(level=1)
brands_merge = pd.merge(brands_trend_12.T.reset_index(level=1), after_con , left_index=True, right_index=True)
brands_merge.drop(columns=[pd.Timestamp('2008-09-01'),pd.Timestamp('2008-10-01'),
                          pd.Timestamp('2008-11-01'),
                          pd.Timestamp('2008-12-01'),
                           pd.Timestamp('2009-01-01'),
                           pd.Timestamp('2009-02-01'),
                           pd.Timestamp('2009-03-01'),
                           pd.Timestamp('2009-04-01'),
                           pd.Timestamp('2009-05-01'),
                           pd.Timestamp('2009-06-01'),
                           pd.Timestamp('2009-07-01'),
                           pd.Timestamp('2009-08-01')],
                           inplace = True
                          )

max_brands = brands_merge[brands_merge['level_1']== 'max'].drop(columns = 'level_1')
median_brands = brands_merge[brands_merge['level_1']== 'median'].drop(columns = 'level_1')
sum_brands = brands_merge[brands_merge['level_1']== 'sum'].drop(columns = 'level_1')
std_brands = brands_merge[brands_merge['level_1']== 'std'].drop(columns = 'level_1')

from datetime import datetime
max_brandinfo = []

for i in range(len(max_brands)):
    for columns_key in range (len(max_brands.columns)):
        if max_brands.columns[columns_key] ==  pd.Timestamp(datetime.strptime(max_brands.iat[i,-1], '%Y-%m-%d').strftime('%Y-%m')):
            max_brandinfo.append(max_brands.iat[i,columns_key])
            
tmp = pd.DataFrame({'brand_max':max_brandinfo})
all_brands = pd.concat([max_brands.iloc[:, -3:-1], tmp], axis = 1)

median_brandinfo = []

for i in range(len(median_brands)):
    for columns_key in range (len(median_brands.columns)):
        if median_brands.columns[columns_key] ==  pd.Timestamp(datetime.strptime(median_brands.iat[i,-1], '%Y-%m-%d').strftime('%Y-%m')):
            #print(median_brands.columns[columns_key] )
            median_brandinfo.append(median_brands.iat[i,columns_key])
            
tmp = pd.DataFrame({'brand_median':median_brandinfo})
tmp_brands = pd.concat([median_brands.iloc[:, -3:-1], tmp], axis = 1)
all_brands = pd.merge(all_brands, tmp_brands, on=['company_id', 'brandall'], how='left')

sum_brandinfo = []

for i in range(len(sum_brands)):
    for columns_key in range (len(sum_brands.columns)):
        if sum_brands.columns[columns_key] ==  pd.Timestamp(datetime.strptime(sum_brands.iat[i,-1], '%Y-%m-%d').strftime('%Y-%m')):
            #print(sum_brands.columns[columns_key] )
            sum_brandinfo.append(sum_brands.iat[i,columns_key])
            
tmp = pd.DataFrame({'brand_sum':sum_brandinfo})
tmp_brands = pd.concat([sum_brands.iloc[:, -3:-1], tmp], axis = 1)
all_brands = pd.merge(all_brands, tmp_brands, on=['company_id', 'brandall'], how='left')

std_brandinfo = []

for i in range(len(std_brands)):
    for columns_key in range (len(std_brands.columns)):
        if std_brands.columns[columns_key] ==  pd.Timestamp(datetime.strptime(std_brands.iat[i,-1], '%Y-%m-%d').strftime('%Y-%m')):
            std_brandinfo.append(std_brands.iat[i,columns_key])
            
tmp = pd.DataFrame({'brand_std':std_brandinfo})
tmp_brands = pd.concat([std_brands.iloc[:, -3:-1], tmp], axis = 1)
all_brands = pd.merge(all_brands, tmp_brands, on=['company_id', 'brandall'], how='left')

tmp  = all_brands[['company_id', 'brand_max', 'brand_median', 'brand_sum', 'brand_std']].groupby(['company_id'], as_index=False).agg('max')
tmp .columns = ['company_id',  '12k_max_brand_max', '12k_max_brand_median', '12k_max_brand_sum', '12k_max_brand_std']

companyinfo = to_get_brands[['company_id', 'company_name'] ]
companyinfo = pd.merge(tmp,companyinfo, on= ['company_id'], how= 'left')
```

```python
companyinfo.drop_duplicates(subset='company_id').reset_index(drop = True)
```
